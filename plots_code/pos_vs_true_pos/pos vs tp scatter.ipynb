```python

import matplotlib as mpl
import matplotlib.pyplot as plt
from adjustText import adjust_text
from pylab import *
from pylab import ylim

mpl.rcParams["figure.dpi"] = 300
```

```python
path_to_save = (
    "/home/krushna/Documents/scDREAMER-experiments/plots_code/pos_vs_true_pos/Figures/"
)
path_all_metrics_excel = "/home/krushna/Downloads/All_metrics_15_Mar_23.xlsx"
sheet_name = "True_pos"
method_color_dct = {
    "scVI": "#28DDED",
    "Harmony": "#ED7A28",
    "Seurat": "#994363",
    "BBKNN": "#B626D3",
    "Scanorama": "#EDBF28",
    "INSCT": "#286CED",
    "LIGER": "#90EE90",
    "fastMNN": "#FFB6C1",
    "iMAP": "#964B00",
    "scDML": "#6F3AF9",
    "scDREAMER": "#086E28",
    "scANVI": "#c5b0d5",
    "scGEN": "#d62829",
    "scDREAMER-Sup": "#113f0a",
}
method_color_dct.update(
    {
        "scDREAMER-Sup-10": method_color_dct["scDREAMER-Sup"],
        "scDREAMER-Sup-20": method_color_dct["scDREAMER-Sup"],
        "scDREAMER-Sup-50": method_color_dct["scDREAMER-Sup"],
        "scANVI-10": method_color_dct["scANVI"],
        "scANVI-20": method_color_dct["scANVI"],
        "scANVI-50": method_color_dct["scANVI"],
        "scGEN-10": method_color_dct["scGEN"],
        "scGEN-20": method_color_dct["scGEN"],
        "scGEN-50": method_color_dct["scGEN"],
    }
)
to_plot_only = ["scVI", "Harmony", "scDREAMER"]  # ,'LIGER']
font_dict = {"family": "Arial", "weight": "bold", "size": 7}
```

```python
# top 3 methods pos vs tp scatter plot
for dataset in ["Pancreas", "Lung", "Immune_Human", "Human_Retina"]:  #'Human_Mouse'
    print("plotting for dataset ", dataset)
    df_dataset = df.loc[df["Dataset"] == dataset, :]
    df_dataset.index = df_dataset["Method"]
    if dataset in ["Human_Retina"]:
        df_dataset = df_dataset.loc[["scDREAMER", "scVI", "scDML"], :]
    elif dataset in ["Pancreas"]:
        df_dataset = df_dataset.loc[["Seurat", "iMAP", "scDREAMER"], :]
    else:
        df_dataset = df_dataset.loc[to_plot_only, :]
    df_dataset["color"] = df_dataset["Method"].replace(method_color_dct)
    #     display(df_dataset)
    plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
    plt.xlabel("Positive cells (%)", fontsize=15, fontname="Arial", fontweight="bold")
    plt.ylabel(
        "True positive cells (%)", fontsize=15, fontname="Arial", fontweight="bold"
    )
    # adjust_text(ts, x=x, y=y)
    plt.subplot(111).spines["right"].set_visible(False)
    plt.subplot(111).spines["top"].set_visible(False)
    plt.subplot(111).get_legend().remove()
    ax = plt.gca()
    for tick in ax.xaxis.get_major_ticks():
        tick.label1.set_fontsize(15)
        tick.label1.set_fontweight("bold")
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontsize(15)
        tick.label1.set_fontweight("bold")
    plt.tight_layout()
    plt.xticks(rotation=75)
    plt.gcf().subplots_adjust(bottom=0.3)

    plt.scatter(
        df_dataset["positive %"], df_dataset["True Pos %"], c=df_dataset["color"]
    )
    ts = [
        plt.text(
            float(df_dataset["positive %"][i]),
            float(df_dataset["True Pos %"][i]),
            txt,
            fontdict=font_dict,
        )
        for i, txt in enumerate(df_dataset["Method"])
    ]
    adjust_text(ts, expand_points=(1.2, 1.2))
    print("ylim", ylim())

    if dataset == "Lung":
        plt.xlim(75, 85)
        plt.ylim(0, 7)
    if dataset == "Immune_Human":
        plt.xlim(88, 92)
        plt.ylim(0, 50)
    if dataset == "Human_Retina":
        plt.xlim(97, 99)
        plt.ylim(0, 80)

    plt.savefig(path_to_save + dataset + "_scatter.png", transparent=True)
    plt.show()
```
