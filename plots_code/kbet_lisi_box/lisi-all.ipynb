```python
def plot_box_multi(x,y,hue,palette,y_label,save=None):
    ax = sns.boxplot(x = x,
                y = y,
                hue = hue,palette=palette)
    plt.xticks(rotation=90)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'family':"Arial",'weight':'bold','size':14})
    ax.get_legend().remove()
    for tick in ax.xaxis.get_major_ticks():
        tick.label1.set_fontsize(14)
        tick.label1.set_fontweight('bold')
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontsize(14)
        tick.label1.set_fontweight('bold')
    ax.set_ylabel(y_label,fontsize = 15, fontname='Arial', fontweight = 'bold')
    ax.set_xlabel('Percentage of missing cell type labels',fontsize = 15, fontname='Arial', fontweight = 'bold')
    ax.spines['top']. set_visible(False)
    ax.spines['right']. set_visible(False)
    if save:
        print('plot saving at : ',save+y_label+'.png')
        plt.savefig(save+y_label+'.png', transparent=True, bbox_inches='tight')
    plt.show()
```

```python
import seaborn as sns
import pandas as pd
plot_save_path_sup = "/home/krushna/Documents/Data_integration/Figures/lisi/lisi_sup/"
base_path = '/home/krushna/Documents/Data_integration/Othermethods/'
datasets = ['Lung','Immune_Human']
for dataset in datasets:
    print(dataset)
    ilisi_paths = [
            "scGen/kbet_lisi/"+dataset+"_ilisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.1_ilisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.2_ilisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.5_ilisi.csv",
            "scanvi/kbet_lisi/"+dataset+"_ilisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.1_ilisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.2_ilisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.5_ilisi.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_ilisi.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_ilisi_NA_0.1.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_ilisi_NA_0.1.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_ilisi_NA_0.1.csv"
                  ]
    clisi_paths = [
            "scGen/kbet_lisi/"+dataset+"_clisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.1_clisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.2_clisi.csv",
            "scGen/NA_102050/"+dataset+"_NA_0.5_clisi.csv",
            "scanvi/kbet_lisi/"+dataset+"_clisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.1_clisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.2_clisi.csv",
            "scanvi/NA_102050/"+dataset+"_NA_0.5_clisi.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_clisi.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_clisi_NA_0.1.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_clisi_NA_0.1.csv",
            "ourmethod/kbet_lisi/supervised/"+dataset+"++_clisi_NA_0.1.csv"
                  ]
    labels_to_plot = ['scGEN','scGEN-10','scGEN-20','scGEN-50','scANVI','scANVI-10','scANVI-20','scANVI-50','scDREAMER-Sup','scDREAMER-Sup-10','scDREAMER-Sup-20','scDREAMER-Sup-50']
    lisi_data = pd.DataFrame([],columns=['iLISI','method','Percentage_wrong'])
    for path,method in zip(ilisi_paths,labels_to_plot):
        temp = pd.DataFrame(np.genfromtxt(base_path+path),columns=['iLISI'])
        temp['method'] = method.rsplit('-',1)[0]
        if method == 'scDREAMER-Sup':
            temp['method'] = method
            temp['Percentage_wrong'] = 0
        elif len(method.split('-'))==1 :
            temp['Percentage_wrong'] = 0
        else:
            temp['Percentage_wrong'] = method.rsplit('-',1)[1]
        lisi_data = pd.concat([lisi_data,temp],axis=0)
    plot_box_multi(x=lisi_data['Percentage_wrong'],y=lisi_data['iLISI'],hue=lisi_data['method'],palette=method_color_dct,y_label='iLISI',
                   save=plot_save_path_sup+dataset+'_sup_ilisi.png')
    
    lisi_data = pd.DataFrame([],columns=['cLISI','method','Percentage_wrong'])
    for path,method in zip(clisi_paths,labels_to_plot):
        temp = pd.DataFrame(np.genfromtxt(base_path+path),columns=['cLISI'])
        temp['method'] = method.rsplit('-',1)[0]
        if method == 'scDREAMER-Sup':
            temp['method'] = method
            temp['Percentage_wrong'] = 0
        elif len(method.split('-'))==1 :
            temp['Percentage_wrong'] = 0
        else:
            temp['Percentage_wrong'] = method.rsplit('-',1)[1]
        lisi_data = pd.concat([lisi_data,temp],axis=0)
    plot_box_multi(x=lisi_data['Percentage_wrong'],y=lisi_data['cLISI'],hue=lisi_data['method'],palette=method_color_dct,y_label='cLISI',
                   save=plot_save_path_sup+dataset+'_sup_clisi.png')
```
