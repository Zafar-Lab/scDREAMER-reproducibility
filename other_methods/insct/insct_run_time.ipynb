```python
import scanpy as sc
from insct.tnn import TNN
import ivis
import numpy as np
```

```python
dir_path = "/home/krushna/Documents/Data_integration/SCRNA_Datasets/All_h5ad/"
def load_data(dataset,batch):
    adata =sc.read_h5ad(dir_path+dataset+'.h5ad')
    sc.pp.filter_genes(adata, min_counts=3)
    adata.layers["counts"] = adata.X.copy()
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    adata.raw = adata
    sc.pp.highly_variable_genes(
            adata,
            flavor="seurat",
            n_top_genes=2000,
            layer="counts",
            batch_key=batch,
            subset=True
    )

    return adata
```

```python
dataset  = 'Human_Mouse_10k'
batch =  'batch'
cell_type = "celltype"
adata = load_data(dataset,batch)
sc.pp.neighbors(adata)
```

```python
import time
X = adata.obsm["X_pca"] 
start_time = time.time()

model = TNN(epochs = 50)
model.fit(X = adata, batch_name=batch)
adata.obsm["X_tnn"] = model.transform(X = adata)

total_time = time.time()-start_time
print("total time taken: ",total_time)
```
